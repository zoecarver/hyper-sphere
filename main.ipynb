```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

%matplotlib notebook
```

## The idea here is that I want to learn how to visualize 4d shapes
---
## To start I made a simple function that generates a sphere
I then added `c` which started as color but turned into the forth dimentions (often refered to as `w`)

```python
def generate_sphere_points(r, center):
    h, k, l = center
    
    xs = []
    ys = []
    zs = []
    cs = []
        
    for a in range(r):
        for b in range (r):
            x = np.sin(a) * np.sin(b) + h
            y = np.sin(a) * np.cos(b) + k
            z = np.sin(r) * np.cos(a) + l
            w = np.sin(r) * np.sin(a) + l
            xs += [x]
            ys += [y]
            zs += [z]
            cs += [w]
        
    return np.array(xs), np.array(ys), np.array(zs), np.array(cs)
```

## Output
Here is what the sphere looks like (technacally a "shadow" of a hyper-shpere) where the forth demention is represeted as color.

```python
fig = plt.figure()
ax = Axes3D(fig)

x, y, z, c = generate_sphere_points(25, (0, 0, 0))

ax.scatter(x, y, z, c=c)
```

# shadows of sphere
here are what the shadows of the sphere looks like (ignore the scaling issues)

Notice how they are all circles

```python
plt.scatter(x, y)
plt.scatter(y, z)
plt.scatter(z, y)
```

# shadow of hyper-sphere
We do the exact same thing except with another dimention

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(x, y, z)
ax.scatter(y, z, c)
ax.scatter(z, c, x)
ax.scatter(c, x, y)
```

# Open up the shadows so they can be seen better

```python
fig1 = plt.figure()
ax1 = Axes3D(fig1)
ax1.scatter(x, y, z)

fig2 = plt.figure()
ax2 = Axes3D(fig2)
ax2.scatter(y, z, c)

fig3 = plt.figure()
ax3 = Axes3D(fig3)
ax3.scatter(z, c, x)

fig4 = plt.figure()
ax4 = Axes3D(fig4)
ax4.scatter(c, x, y)
```
